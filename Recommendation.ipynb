```python
import pandas as pd
```

```python
df = pd.read_csv('file.tsv', sep='\t', names=['user id', 'item_id', 'rating', 'timestamp'])
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles.csv')
```

```python
movie_titles.head()
```

```python
data = pd.merge(df, movie_titles, on='item_id') #on, how=inner, left, right, outer
```

```python
data.head()
```

```python
# Calculating the mean rating of all movies

data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
# calculate count rating of all movies

data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
# creating dataframe with rating cunt values

ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
ratings['number of ratings'] = pd.DataFrame(data.groupby('title')['rating'].count())

ratings.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(ratings['number of ratings'], bins=70)
```

```python
plt.hist(ratings['rating'], bins=70)
```

```python
movie_mat = data.pivot_table(index='user id', columns='title', values='rating')
```

```python
movie_mat.head()
```

```python
movie_mat.fillna(0, inplace=True)
```

```python
# analysing correlation with similar movies

starwar_movie_rating = movie_mat['Star Wars (1977)']
liar_movie_rating = movie_mat['Liar Liar (1997)']
```

```python
similar_to_starwars = movie_mat.corrwith(starwar_movie_rating)
```

```python
similar_to_lier = movie_mat.corrwith(liar_movie_rating)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head()
```

```python
corr_starwars = corr_starwars.join(ratings['number of ratings'])
```

```python
corr_starwars.head()
```
