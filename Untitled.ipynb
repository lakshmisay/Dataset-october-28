```python
import pandas as pd
```

```python
df = pd.read_excel('Online_Retail.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
basket = df[df['Country']=='France'].groupby(['InvoiceNo','Description' ])['Quantity'].sum().unstack().fillna(0)
```

```python
basket
```

```python
def encode(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
basket_sets = basket.applymap(encode)
```

```python
basket_sets.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemset = apriori(basket_sets, min_support=0.04, use_colnames=True)
```

```python
rules = association_rules(frequent_itemset, metric='lift', min_threshold=1)
```

```python
rules.head()
```

```python
rules[rules['confidence']>0.8]
```

```python
Antecedent= item is purchased
consequent =  chance of going to purchased

support(A) = product A / total number of product = likelihood of purchasing product xx (freuently)

support(C) = product C / total number of product

Support = Product A & C/ Total number of poroduct

confidence =  product A & C /support (A)

Lift = Support / Support(A)*Support(C)
```
